```python
import pandas as pd
```

```python
df = pd.read_csv("/content/taylor_swift_spotify.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['release_date']= pd.to_datetime(df['release_date'])
```

```python
df.info()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.head()
```

```python
album_popularity = (
    df.groupby('album')['popularity']
    .mean()
    .reset_index()
    .sort_values(by='popularity', ascending=False)
    .reset_index(drop=True)
)
album_popularity.index += 1
print(album_popularity)
```

```python
total_music_by_album = (
    df.groupby('album')['name']
    .count()
    .reset_index(name='music_count')
    .sort_values(by='music_count', ascending=False)
)
total_music_by_album.index += 1
print(total_music_by_album)
```

```python
top10 = (
    df[['name', 'album', 'popularity']]
    .sort_values(by='popularity', ascending=False)
    .reset_index(drop=True)
    .head(10)
)
top10.index += 1
```

```python
top10
```

```python
longest_5_songs = (
    df[['name', 'duration_ms']]
    .sort_values(by='duration_ms', ascending=False)
    .drop_duplicates(subset='name')
    .head(5)
    .reset_index(drop=True)
)

longest_5_songs['duration_min'] = (
    longest_5_songs['duration_ms'] / 60000
).round(2)

longest_5_songs.index += 1

print(longest_5_songs)
```

```python
print(longest_5_songs)
```

```python
popular_songs_classification = (
    df.groupby(['name','danceability','energy'])['popularity']
    .mean()
    .reset_index()
    .sort_values(by=['popularity'], ascending=False)
    .reset_index(drop=True)
    .head(10)
)
popular_songs_classification.index +=1
```

```python
print(popular_songs_classification)
```

```python
album_popularity.to_csv('album_popularity.csv', index=False)
total_music_by_album.to_csv('total_music_by_album.csv', index=False)
top10.to_csv('top10.csv', index=False)
longest_5_songs.to_csv('longest_5_songs.csv', index=False)
popular_songs_classification.to_csv('popular_songs_classification.csv', index=False)
```
